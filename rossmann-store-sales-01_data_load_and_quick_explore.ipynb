```python
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Google Driveのマウント
drive.mount('/content/drive/')

# データの読み込み
def load_data(train_path, store_path):
    train = pd.read_csv(train_path).rename(columns=lambda x: x.strip())
    store = pd.read_csv(store_path).rename(columns=lambda x: x.strip())
    return train.merge(store, on="Store", how="left")

train_path = "/content/drive/MyDrive/Colab Notebooks/standard/rossmann-store-sales/data/train.csv"
store_path = "/content/drive/MyDrive/Colab Notebooks/standard/rossmann-store-sales/data/store.csv"
data = load_data(train_path, store_path)

# データの基本情報
def basic_info(df):
    print("\n### Basic Information ###")
    print(df.info())
    print("\n### Sample Data ###")
    print(df.head())
    print("\n### Missing Values ###")
    print(df.isnull().sum())
    print("\n### Descriptive Statistics ###")
    print(df.describe())

# メイン処理
def main():
    basic_info(data)

if __name__ == "__main__":
    main()
```
